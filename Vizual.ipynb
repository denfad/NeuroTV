```python
import copy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from scipy.stats import pearsonr
import warnings
warnings.filterwarnings("ignore")
sns.set(context='notebook', style='whitegrid', palette='deep', 
        font='sans-serif', font_scale=1, color_codes=False, rc=None)
```

```python
df_data = pd.read_csv('data.csv', sep=';')
df_params = pd.read_csv('params.csv', sep=';')
df_regions = pd.read_csv('regions.csv', sep=';')
df_train = pd.read_csv('train.csv')
df_train.drop('Unnamed: 0', axis=1)
```

```python
df_train['Выпуск специалистов высшими учебными заведениями, %'] = 100 * df_train['Выпуск специалистов высшими учебными заведениями, тыс. чел.'] / df_train['Численность населения региона; на конец года; тыс. чел.']
df_train['ВВП региона на душу населения'] = df_train['Валовой региональный продукт; млн руб.'] / df_train['Численность населения региона; на конец года; тыс. чел.']
df_train['Расходы на общегосударственные вопросы, %'] = 100 * df_train['Расходы на общегосударственные вопросы, млн. руб.'] / df_train['Расходы региональных бюджетов общие, млн.руб.']
df_train['Расходы региональных бюджетов по статье национальная экономика, %'] = 100 * df_train['Расходы региональных бюджетов по статье национальная экономика'] / df_train['Расходы региональных бюджетов общие, млн.руб.']
df_train['Затраты на информационные и коммуникационные технологии (ИКТ), %'] = 100 * df_train['Затраты на информационные и коммуникационные технологии (ИКТ), млн. руб'] / df_train['Расходы региональных бюджетов общие, млн.руб.']
df_train['Затраты на технологические инновации, %'] = 100 * df_train['Затраты на технологические инновации, млн. руб.'] / df_train['Расходы региональных бюджетов общие, млн.руб.']
df_train['Расходы консолидированных бюджетов субъектов Российской Федерации на социально-культурные мероприятия, %'] = 100 * df_train['Расходы консолидированных бюджетов субъектов Российской Федерации на социально-культурные мероприятия, млн руб.'] / df_train['Расходы региональных бюджетов общие, млн.руб.']
df_train['Численность работников органов исполнительной власти субъектов РФ и органов местного самоуправления, %'] = 100 * df_train['Численность работников органов исполнительной власти субъектов РФ и органов местного самоуправления; чел.'] / df_train['Численность населения региона; на конец года; тыс. чел.']
```

```python
df_train = df_train[df_train['Площадь территории по субъектам Российской Федерации; кв. км.'] != 0]
df_train = df_train[df_train['ВВП региона на душу населения'].notna()]
```

```python
param = 'Комфортность климата'
k = 10
corrmat = df_train.corr()
cols = corrmat.nlargest(k, param)[param].index
cm = np.corrcoef(df_train[cols].values.T)
sns.set(font_scale=1.25)
plt.figure(figsize = (10, 10))
hm = sns.heatmap(cm, cbar=True, annot=True, square=True,
                fmt='.3f', annot_kws={'size': 10}, cmap='RdYlGn',
                center=0, vmax=1, vmin=-1,
                yticklabels=cols.values, xticklabels=cols.values)
```

```python
param = 'Комфортность климата'
k = 10
h = [-1]
h.extend(list(range(k)))
corrmat = df_train.corr()
cols = corrmat.sort_values(by=param, ascending=True)[param][h].index
cm = np.corrcoef(df_train[cols].values.T)
sns.set(font_scale=1.25)
plt.figure(figsize = (23, 23))
hm = sns.heatmap(cm, cbar=True, annot=True, square=True,
                fmt='.2f', annot_kws={'size': 15}, cmap='RdYlGn',
                center=0, vmax=1, vmin=-1,
                yticklabels=cols.values, xticklabels=cols.values)
```
