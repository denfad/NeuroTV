```python
import copy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
import warnings
warnings.filterwarnings("ignore")
sns.set(context='notebook', style='whitegrid', palette='deep', 
        font='sans-serif', font_scale=1, color_codes=False, rc=None)
```

```python
df_data = pd.read_csv('data.csv', sep=';')
df_params = pd.read_csv('params.csv', sep=';')
df_regions = pd.read_csv('regions.csv', sep=';')
```

```python
#создание норм таблицы


df_test = pd.DataFrame()
df_test['Округ'] = "круг"
df_test['Регион'] = "регион"
df_test['Год'] = 2007
for x in df_params["name"].unique():
    df_test[x] = 5.7
full = []
for region_id in df_regions[df_regions["id"] > 10]["id"].unique():
    for year in df_data[df_data["region_id"] == region_id]["year_id"].unique():
        row = []
        row.append(df_regions[df_regions["id"] == df_regions[df_regions["id"] == region_id]["upid"].values[0]]["name"].values[0])
        row.append(df_regions[df_regions.id == region_id]["name"].values[0])
        row.append(year)
       
        for param_id in df_params["id"].unique():
            if param_id in df_data[(df_data["region_id"] == region_id) & (df_data["year_id"] == year)]["param_id"].unique():
                row.append(df_data[(df_data["region_id"] == region_id) & (df_data["year_id"] == year) & (df_data["param_id"] == param_id)]["value"].values[0])
            else:
                row.append(0)
        full.append(row)
        

df_test.loc[:] = full
df_test.to_csv("train", sep='\t')


```

```python
#итоговая норм таблица
df_train = df_test.copy()
```
